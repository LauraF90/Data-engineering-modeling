# The CSV co2final.csv is the produced file of these scripts.

```python
# query the data from 1-1-2018 to 31-12-2018
client = EntsoePandasClient(api_key='5aa63adb-f9f7-47aa-8053-5cf531a3743f')
start = pd.Timestamp('20180101', tz='Europe/Berlin')
end = pd.Timestamp('20190101', tz='Europe/Berlin')
country_code = 'DE'
ts2 = client.query_generation(country_code, start=start, end=end, psr_type=None)
ts2
```

```python
# The values of these three columns are always Nan or zero so there is no need for them and can be dropped
# from year 2015 to 2017
# ts[ts['Fossil Oil shale'] > 0]
# to check how many null values in each column
# ts.isnull().sum()
df = df.drop('Fossil Oil shale' , axis=1)
df = df.drop('Fossil Peat' , axis=1)
df = df.drop('Marine' , axis=1)
```

```python
# function to multiply all values to 0.25 to make the generation MWh and then to 1000 to get it from MWh to kWh
def kwh (ts):
    ts_mwh= ts.multiply(0.25)
    ts_kwh = ts_mwh * 1000
    return ts_kwh
```

```python
# function to group the data frame by date and hour as multiple index and add the columns for the day of the week , day of the month , month , date , hour
# then it creates the column Generation which is the sum of all the generation
# the oil is not included
def toGroup (ts):
    ts['hour']=ts.index.hour
    ts['date']=ts.index.date
    ts['weekDay']=ts.index.dayofweek
    ts['monthDay']=ts.index.day
    ts['month']=ts.index.month
    #ts_kwh_group=ts.groupby(['date','hour']).sum()
    ts_kwh_group=ts.groupby(['date','hour']).agg({'Biomass':'sum', 'Fossil Brown coal/Lignite':'sum', 'Fossil Coal-derived gas':'sum','Fossil Gas':'sum', 'Fossil Hard coal':'sum' , 'Geothermal':'sum' , 'Hydro Pumped Storage':'sum','Hydro Run-of-river and poundage':'sum','Hydro Water Reservoir':'sum','Nuclear':'sum','Other':'sum' ,'Other renewable':'sum' , 'Solar':'sum', 'Waste':'sum', 'Wind Offshore':'sum' ,'Wind Onshore':'sum' ,'weekDay':'max', 'monthDay':'max' , 'month':'max'})
    ts_kwh_group['Generation'] = ts_kwh_group.sum(axis=1)
    return ts_kwh_group
```

```python
# function to group the data frame by date and hour as multiple index and add the columns for the day of the week , day of the month , month , date , hour
# then it creates the column Generation which is the sum of all the generation
# the oil is included
def toGroup2 (ts):
    ts['hour']=ts.index.hour
    ts['date']=ts.index.date
    ts['weekDay']=ts.index.dayofweek
    ts['monthDay']=ts.index.day
    ts['month']=ts.index.month
    #ts_kwh_group=ts.groupby(['date','hour']).sum()
    ts_kwh_group=ts.groupby(['date','hour']).agg({'Biomass':'sum', 'Fossil Brown coal/Lignite':'sum', 'Fossil Coal-derived gas':'sum','Fossil Gas':'sum', 'Fossil Hard coal':'sum' , 'Fossil Oil':'sum', 'Geothermal':'sum' , 'Hydro Pumped Storage':'sum','Hydro Run-of-river and poundage':'sum','Hydro Water Reservoir':'sum','Nuclear':'sum','Other':'sum' ,'Other renewable':'sum' , 'Solar':'sum', 'Waste':'sum', 'Wind Offshore':'sum' ,'Wind Onshore':'sum' ,'weekDay':'max', 'monthDay':'max' , 'month':'max'})
    ts_kwh_group['Generation'] = ts_kwh_group.sum(axis=1)
    return ts_kwh_group
```

```python
# function to calculate the Co2 emission for the years between 2015 to 2017 and returns the value as a column
# GHG = [71,820,800,400,800,45,34,4,9,11,247,71,43,690,9,8]
def Co2_cal (Bi,FB,FC,FG,FH,Ge,HP,HR,HW,Nu,Ot,OR,So,Wa,WOf,WOn) :
    
    return ((Bi*71)+(FB*820)+(FC*800)+(FG*400)+(FH*800)+(Ge*45)+(HP*34)+(HR*4)+(HW*9)+(Nu*11)+(Ot*247)+(OR*71)+(So*43)+(Wa*690)+(WOf*9)+(WOn*8))/(Bi+FB+FC+FG+FH+Ge+HP+HR+HW+Nu+Ot+OR+So+Wa+WOf+WOn)
Co2_cal(100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100)
```

```python
# function to calculate the Co2 emission for the years between 2018 and returns the value as a column
# GHG = [71,820,800,400,800,520,45,34,4,9,11,247,71,43,690,9,8]
def Co2_cal2 (Bi,FB,FC,FG,FH,FO,Ge,HP,HR,HW,Nu,Ot,OR,So,Wa,WOf,WOn) :
    
    return ((Bi*71)+(FB*820)+(FC*800)+(FG*400)+(FH*800)+(FO*520)+(Ge*45)+(HP*34)+(HR*4)+(HW*9)+(Nu*11)+(Ot*247)+(OR*71)+(So*43)+(Wa*690)+(WOf*9)+(WOn*8))/(Bi+FB+FC+FG+FH+FO+Ge+HP+HR+HW+Nu+Ot+OR+So+Wa+WOf+WOn)
```

```python
# script to call the function in order to prepare the data for 2015 , 2016 , 2017
ts_kwh = kwh(df)
ts_kwh_group = toGroup(ts_kwh)
ts_kwh_group['Co2_emission'] = Co2_cal(ts_kwh_group['Biomass'] , ts_kwh_group['Fossil Brown coal/Lignite'] , ts_kwh_group['Fossil Coal-derived gas'] ,
                            ts_kwh_group['Fossil Gas'] , ts_kwh_group['Fossil Hard coal'] , ts_kwh_group['Geothermal'] , ts_kwh_group['Hydro Pumped Storage'],
                            ts_kwh_group['Hydro Run-of-river and poundage'] , ts_kwh_group['Hydro Water Reservoir'] , ts_kwh_group['Nuclear'],
                            ts_kwh_group['Other'] , ts_kwh_group['Other renewable'] , ts_kwh_group ['Solar'] , ts_kwh_group['Waste'] ,
                            ts_kwh_group['Wind Offshore'] , ts_kwh_group['Wind Onshore'])
ts_kwh_group
# delete the null rows which equal to 152 row.
co2_2017 = ts_kwh_group.dropna(axis='index',how ='any')
```

```python
# script to call the function in order to prepare the data for 2018
ts_kwh2 = kwh(ts2)
ts_kwh_group2 = toGroup2(ts_kwh2)
ts_kwh_group2['Co2_emission'] = Co2_cal2(ts_kwh_group2['Biomass'] , ts_kwh_group2['Fossil Brown coal/Lignite'] , ts_kwh_group2['Fossil Coal-derived gas'] ,
                            ts_kwh_group2['Fossil Gas'] , ts_kwh_group2['Fossil Hard coal'] ,ts_kwh_group2['Fossil Oil'], ts_kwh_group2['Geothermal'] , ts_kwh_group2['Hydro Pumped Storage'],
                            ts_kwh_group2['Hydro Run-of-river and poundage'] , ts_kwh_group2['Hydro Water Reservoir'] , ts_kwh_group2['Nuclear'],
                            ts_kwh_group2['Other'] , ts_kwh_group2['Other renewable'] , ts_kwh_group2['Solar'] , ts_kwh_group2['Waste'] ,
                            ts_kwh_group2['Wind Offshore'] , ts_kwh_group2['Wind Onshore'])
ts_kwh_group2
# delete the null rows and the are around 300 row
co2_2018 = ts_kwh_group2.dropna(axis='index',how ='any')
```

```python
# dropping the not needed columns 
# because we are interested only in total generation , solar , wind and timing data
final_2017 = co2_2017.drop(['Biomass','Fossil Brown coal/Lignite','Fossil Coal-derived gas','Fossil Gas','Fossil Hard coal',
                           'Geothermal','Hydro Pumped Storage','Hydro Run-of-river and poundage','Hydro Water Reservoir',
                           'Nuclear','Other','Other renewable','Waste'] , axis=1)
final_2018 = co2_2018.drop(['Biomass','Fossil Brown coal/Lignite','Fossil Coal-derived gas','Fossil Gas','Fossil Hard coal',
                           'Geothermal','Fossil Oil','Hydro Pumped Storage','Hydro Run-of-river and poundage','Hydro Water Reservoir',
                           'Nuclear','Other','Other renewable','Waste'] , axis=1)
```

```python
# concatinating the two finals since they have the same columns now
final = pd.concat([final_2017,final_2018])
final.to_csv('finalCo2.csv')
```
